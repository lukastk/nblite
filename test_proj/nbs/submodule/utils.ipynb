# Submodule Utils

This notebook demonstrates exporting to a submodule and importing from other modules.

```python
#|default_exp submodule.utils
```

```python
#|export
from my_lib.core import greet, add
```

```python
#|export
def greet_multiple(names: list[str]) -> list[str]:
    """Greet multiple people.
    
    Args:
        names: List of names to greet.
        
    Returns:
        List of greeting strings.
    """
    return [greet(name) for name in names]
```

```python
# Test the function
greet_multiple(["Alice", "Bob", "Charlie"])
```

```python
#|export
def sum_list(numbers: list[int]) -> int:
    """Sum a list of numbers using the add function.
    
    Args:
        numbers: List of integers to sum.
        
    Returns:
        Sum of all numbers.
    """
    result = 0
    for num in numbers:
        result = add(result, num)
    return result
```

```python
# Test sum_list
sum_list([1, 2, 3, 4, 5])
```
