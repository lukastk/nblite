# Utils Module

A simple module to demonstrate the extension hooks.

```python
#|default_exp utils
```

```python
#|export
def greet(name: str) -> str:
    """Greet someone."""
    return f"Hello, {name}!"
```

```python
#|export
def add(a: int, b: int) -> int:
    """Add two numbers."""
    return a + b
```

```python
# Test
print(greet("World"))
print(add(2, 3))
```
