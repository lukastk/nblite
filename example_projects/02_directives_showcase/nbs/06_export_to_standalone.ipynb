# Export To Without Default Export

This notebook demonstrates using `#|export_to` WITHOUT a `#|default_exp` directive.

This is useful when you want a notebook to contribute to one or more modules
without having its own "default" module.

## No default_exp!

Notice there's no `#|default_exp` in this notebook. All exports use `#|export_to`.

```python
# |export_to standalone_utils
def helper_one():
    """A helper function exported to standalone_utils module."""
    return "helper one"
```

```python
# |export_to standalone_utils
def helper_two():
    """Another helper exported to the same module."""
    return "helper two"
```

```python
# |export_to separate_module
def from_separate():
    """This goes to a completely separate module."""
    return "from separate"
```

## Testing

Since this notebook has no `#|default_exp`, regular `#|export` would have
no target module (and would be ignored). Only `#|export_to` cells are exported.

```python
# Test the functions (these are not exported, just for testing)
print(helper_one())
print(helper_two())
print(from_separate())
```
