# Utils Module

This notebook demonstrates export modes by exporting to two different module locations.

```python
#|default_exp utils
```

```python
#|export
def format_name(first: str, last: str) -> str:
    """Format a full name from first and last name."""
    return f"{first} {last}"
```

```python
#|export
def calculate_area(width: float, height: float) -> float:
    """Calculate area of a rectangle."""
    return width * height
```

```python
# Test the functions (not exported)
print(format_name("John", "Doe"))
print(calculate_area(5, 10))
```
