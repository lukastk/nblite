# Export To Directive

This notebook demonstrates the `#|export_to` directive.

**Note**: The `#|export_to` directive marks cells for export similar to `#|export`,
with additional metadata for module path and ordering. The actual behavior depends
on how your export pipeline is configured.

```python
# |default_exp export_to_demo
```

## Basic Export

Regular `#|export` goes to the default module.

```python
# |export
def regular_export():
    """Exported with standard #|export directive."""
    return "regular"
```

## Export To

The `#|export_to` directive can specify a module path and optional order.
Format: `#|export_to module.path [order]`

```python
# |export_to export_to_demo
def explicit_module():
    """Explicitly specifies the target module."""
    return "explicit"
```

```python
# |export_to export_to_demo 50
def with_order():
    """Specifies module and order (lower = earlier in file)."""
    return "ordered"
```

```python
# Test the functions
print(regular_export())
print(explicit_module())
print(with_order())
```
